```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('../Work/Nov/5903 15.xlsx')
```

```python
def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d})".format(pct, absolute)
plt.style.use('seaborn-talk')
```

```python
my_labels = df['Gender'].value_counts().index.to_list()
data = df['Gender'].value_counts()
colors = plt.get_cmap('jet')(np.linspace(0.2, 0.7, len(my_labels)))
wedge_properties = {"edgecolor":"k",'linewidth': 2}


plt.style.use('default')
plt.figure(figsize = (7,7))
plt.pie(df['Gender'].value_counts(),
        labels=my_labels,
        colors=colors,
        autopct=lambda pct: func(pct, data),
        shadow=True,
        startangle=60,
        wedgeprops=wedge_properties,
        textprops={'fontsize': 16})
plt.title('Gender Chart')
plt.legend(title="Gender",
           loc ="lower left")
```

```python
my_labels = df['Age'].value_counts().index.to_list()
data = df['Age'].value_counts()
colors = plt.get_cmap('jet')(np.linspace(0.2, 0.7, len(my_labels)))
wedge_properties = {"edgecolor":"k",'linewidth': 2}


plt.figure(figsize = (7,7))
plt.pie(df['Age'].value_counts(),
        labels=my_labels,
        colors=colors,
        autopct=lambda pct: func(pct, data),
        shadow=True,
        startangle=60,
        wedgeprops=wedge_properties,
        textprops={'fontsize': 16})
plt.title('Age Chart')
plt.legend(title="Age",
           loc ="lower left")
```

```python
my_labels = df['Disability'].value_counts().index.to_list()
data = df['Disability'].value_counts()
bar_colors = ['blue', 'gray', 'gray']
plt.figure(figsize = (7,7))

barplot = plt.bar(x=my_labels,height=data,color=bar_colors)
plt.bar_label(barplot,labels=data)
plt.show()
```
