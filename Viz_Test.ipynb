```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
month = ["May", "June", "July", "August", "September", "October"]
services_provided = {
    'Information Sharing': [426, 376, 440, 671, 592, 437],
    'PSS': [322,415,575,577,391,281],
    'External Referral': [40, 150, 209, 404, 480, 542],
    'Other': [417,433,500,589,295,195],
}

fig, ax = plt.subplots()
ax.stackplot(month, services_provided.values(),
             labels=services_provided.keys(), alpha=0.8)
plt.style.use('seaborn-talk')
plt.figure(figsize = (15,10))
ax.legend(loc='upper left')
ax.set_title('Intersos Services Provided')
ax.set_xlabel('month')
ax.set_ylabel('Number of services')

plt.show()
```

```python
labels = ['Airlift programme', 'Cultural mediation', 'External referral', 'Communication', 'Information sharing', 'Internal referral','Livelihood support','PSS']
men_means = [1, 3, 79, 1, 39,18,1,23]
women_means = [0,4,79,7,82,25,0,32]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, 
    men_means, 
    width, 
    label='Men',
    color='royalblue')
rects2 = ax.bar(x + width/2,
    women_means, 
    width, 
    label='Women',
    color='red')

plt.figure(figsize = (20,15))
ax.set_ylabel('Services')
ax.set_title('Services by group and gender')
ax.set_xticks(x, 
    labels,
    rotation=25)
ax.legend()

ax.bar_label(rects1, 
    padding=3)
ax.bar_label(rects2, 
    padding=3)

fig.tight_layout()

plt.show()
```
