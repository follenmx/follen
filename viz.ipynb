```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../Work/Sept/08/5903 weekly.xlsx')
```

```python
def make_autopct(values): #Creates labels on the pie chart as values
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_autopct


df['Type of service provided/Tipul serviciului oferit'] = [x.split('/')[0] for x in df['Type of service provided/Tipul serviciului oferit']] #Removes everything after /
```

```python
my_labels = np.array(df['Gender/Sex'].unique())
colors = plt.get_cmap('jet')(np.linspace(0.2, 0.7, len(my_labels)))
wedge_properties = {"edgecolor":"k",'linewidth': 2}


plt.figure(figsize = (10,5))
plt.pie(df['Gender/Sex'].value_counts(sort=False),
        labels=my_labels,
        colors=colors,
        autopct=make_autopct(df['Gender/Sex'].value_counts(sort=False)),
        shadow=True,
        startangle=90,
        wedgeprops=wedge_properties,
        textprops={'fontsize': 16})
plt.title('Gender Chart')
plt.legend(title="Gender",
           loc ="lower left")
```

```python
q = list(df['Type of service provided/Tipul serviciului oferit'].value_counts())


font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 13}


def addlabels():  #Adds labels on the chart
    for i in range(len(q)):
        plt.text(q[i]+0.2,i,q[i],**font)


x = df['Type of service provided/Tipul serviciului oferit'].value_counts().index.to_list()
y = df['Type of service provided/Tipul serviciului oferit'].value_counts()
```

```python
plt.figure(figsize = (16,9))
plt.barh(x, y)
addlabels()
plt.title('Type of Services Chart',
          loc="left",
          fontstyle='italic',
          fontsize=20)
```

```python
dl = []
i = 47
while i < 51:
    dl.append(int(df.iloc[:, [i]].sum()))
    i +=1


def addlabels2(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]+0.3, y[i], ha = 'center', **font)


data = {'Services available': dl[0], 'EU flights': dl[1], 'Internet access': dl[2], 'Asylum procedures': dl[3]}
sorted_data = {}
sorted_keys = sorted(data,
                     key=data.get,
                     reverse=True)  


for w in sorted_keys: #Sorting the dictionary by values
    sorted_data[w] = data[w]
    

names = list(sorted_data.keys())
values = list(sorted_data.values())
```

```python
plt.figure(figsize = (16,9))
plt.bar(names, values)
addlabels2(names,values)
plt.title('Information sharing in-depth chart',
          loc="left",
          fontstyle='italic',
          fontsize=20)
```
